```python
class Temperature:

    def __init__(self, celsius):
        self.celsius = celsius

    def to_fahrenheit(self):
        return (self.celsius * 9/5) + 32

    def to_kelvin(self):
        return self.celsius + 273.15


t = Temperature(25)

print("Temperature in Celsius:", t.celsius)
print("Temperature in Fahrenheit:", t.to_fahrenheit())
print("Temperature in Kelvin:", t.to_kelvin())
```

```python
class Interest:

    def __init__(self, principal, rate, time):
        self.principal = principal
        self.rate = rate
        self.time = time

    def simple_interest(self):
        return (self.principal * self.rate * self.time) / 100

    def compound_interest(self):
        amount = self.principal * (1 + self.rate / 100) ** self.time
        return amount - self.principal


i = Interest(10000, 10, 2)

print("Simple Interest:", i.simple_interest())
print("Compound Interest:", i.compound_interest())
```

```python
class CoffeeMachine:

    def __init__(self, water, coffee, milk):
        self.water = water
        self.coffee = coffee
        self.milk = milk

    def make_latte(self):

        if self.water >= 200 and self.coffee >= 20 and self.milk >= 150:

            self.water -= 200
            self.coffee -= 20
            self.milk -= 150

            print("Latte is ready!")

        else:
            print("Not enough resources to make latte.")


machine = CoffeeMachine(300, 100, 200)

machine.make_latte()

print("Remaining Water:", machine.water)
print("Remaining Coffee:", machine.coffee)
print("Remaining Milk:", machine.milk)
```

```python
class Vehicle:

    def __init__(self, name, max_speed):
        self.name = name
        self.max_speed = max_speed

    def display(self):
        print("Vehicle Name:", self.name)
        print("Maximum Speed:", self.max_speed)


class Bus(Vehicle):

    def __init__(self, name, max_speed, seating_capacity):
        super().__init__(name, max_speed)
        self.seating_capacity = seating_capacity

    def reservation(self):
        print("Reservation is available for today.")
        print("Seating Capacity:", self.seating_capacity)


b = Bus("Volvo Bus", 120, 50)

b.display()
b.reservation()
```
